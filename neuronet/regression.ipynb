##### Нейросетевой подход. Допустим у нас есть датасет из 13 картинок, в котором размечаны иконки инстаграмма. Каждая иконка отмечана пятью точками, которые и будут выходом нейросети.

```python
import numpy as np
import cv2
import json
import pandas as pd
import os

from tensorflow import keras
from tensorflow.keras import layers
```

```python
Y = []
X = []

for file in os.listdir("train"):
    points = []
    if 'json' in file:
        with open("train//" + file, "r") as read_file:
            data = json.load(read_file)
            
        for label in data['shapes']:
            for x, y in label['points']:
                points.append(x)
                points.append(y)
        Y.append(points) 
        
    if 'jpg' in file or 'png' in file or 'jpeg' in file:
        img = cv2.imread("train//" + file, 0)
        X.append(img)
```

```python
X = np.array(X)
Y = np.array(Y)
print(X.shape)
print(Y.shape)
```

```python
input_shape = (220, 100, 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.01, random_state=42)
```

```python
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")
```

```python
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(10, activation="relu"),
    ]
)

model.summary()
```

```python
batch_size = 1
epochs = 1000

model.compile(optimizer='rmsprop',
              loss='mse',
              metrics=['mae'])

model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
